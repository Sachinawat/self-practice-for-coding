```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Assuming 'data' is your DataFrame already loaded

# Drop 'Date' column if it exists
if 'Date' in data.columns:
    data.drop('Date', axis=1, inplace=True)

# Calculate correlation of all columns with 'high' column
correlation_with_high = data.corr()['high'].sort_values(ascending=False)

print("=" * 60)
print("CORRELATION OF ALL COLUMNS WITH 'HIGH' (Descending Order)")
print("=" * 60)
print(correlation_with_high)
print("\n")

# Display top 10 most correlated features (excluding 'high' itself)
print("=" * 60)
print("TOP 10 MOST POSITIVELY CORRELATED WITH 'HIGH'")
print("=" * 60)
top_positive = correlation_with_high[1:11]  # Skip 'high' itself
print(top_positive)
print("\n")

# Display top 10 most negatively correlated features
print("=" * 60)
print("TOP 10 MOST NEGATIVELY CORRELATED WITH 'HIGH'")
print("=" * 60)
bottom_negative = correlation_with_high[-10:]
print(bottom_negative)
print("\n")

# ============================================================
# MAIN VISUALIZATION: Descending order with color gradient
# ============================================================
plt.figure(figsize=(14, 12))

# Remove self-correlation with 'high'
correlation_sorted = correlation_with_high.drop('high')

# Create color map based on correlation values
colors = plt.cm.RdYlGn(np.linspace(0.2, 0.9, len(correlation_sorted)))
colors_reversed = colors[::-1]  # Reverse to match descending order

# Create horizontal bar chart
bars = plt.barh(range(len(correlation_sorted)), correlation_sorted.values, color=colors_reversed, edgecolor='black', linewidth=0.5)

# Add value labels on bars
for i, (value, bar) in enumerate(zip(correlation_sorted.values, bars)):
    plt.text(value + 0.01 if value >= 0 else value - 0.01, i, f'{value:.3f}', 
             va='center', ha='left' if value >= 0 else 'right', fontsize=9, fontweight='bold')

plt.yticks(range(len(correlation_sorted)), correlation_sorted.index, fontsize=11)
plt.xlabel('Correlation Coefficient with "high"', fontsize=13, fontweight='bold')
plt.title('Feature Correlation with "high" (Descending Order)\nGreen = Strong Positive | Red = Negative', 
          fontsize=15, fontweight='bold', pad=20)
plt.axvline(x=0, color='black', linestyle='-', linewidth=1.5)
plt.grid(axis='x', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

# ============================================================
# ALTERNATIVE VIEW: Heatmap with descending order
# ============================================================
plt.figure(figsize=(10, 14))

correlation_matrix = correlation_sorted.to_frame()
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlGn', center=0, 
            linewidths=2, linecolor='black', cbar_kws={"shrink": 0.8}, 
            annot_kws={"size": 11, "fontweight": "bold"}, fmt='.3f', 
            vmin=-1, vmax=1)

plt.title('Correlation Heatmap: Features vs "high" (Descending Order)', 
          fontsize=15, fontweight='bold', pad=15)
plt.ylabel('Features', fontsize=13, fontweight='bold')
plt.xlabel('Correlation with "high"', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

# ============================================================
# ENHANCED VIEW: Separated Positive and Negative Correlations
# ============================================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 10))

# Positive correlations
positive_corr = correlation_sorted[correlation_sorted > 0]
colors_positive = plt.cm.Greens(np.linspace(0.4, 0.9, len(positive_corr)))[::-1]
ax1.barh(range(len(positive_corr)), positive_corr.values, color=colors_positive, edgecolor='black', linewidth=0.5)
ax1.set_yticks(range(len(positive_corr)))
ax1.set_yticklabels(positive_corr.index, fontsize=10)
ax1.set_xlabel('Correlation Coefficient', fontsize=12, fontweight='bold')
ax1.set_title('Positive Correlations with "high"', fontsize=13, fontweight='bold')
ax1.grid(axis='x', alpha=0.3)

# Add value labels
for i, value in enumerate(positive_corr.values):
    ax1.text(value + 0.01, i, f'{value:.3f}', va='center', fontsize=9, fontweight='bold')

# Negative correlations
negative_corr = correlation_sorted[correlation_sorted < 0].sort_values(ascending=True)
if len(negative_corr) > 0:
    colors_negative = plt.cm.Reds(np.linspace(0.4, 0.9, len(negative_corr)))[::-1]
    ax2.barh(range(len(negative_corr)), negative_corr.values, color=colors_negative, edgecolor='black', linewidth=0.5)
    ax2.set_yticks(range(len(negative_corr)))
    ax2.set_yticklabels(negative_corr.index, fontsize=10)
    ax2.set_xlabel('Correlation Coefficient', fontsize=12, fontweight='bold')
    ax2.set_title('Negative Correlations with "high"', fontsize=13, fontweight='bold')
    ax2.grid(axis='x', alpha=0.3)
    
    # Add value labels
    for i, value in enumerate(negative_corr.values):
        ax2.text(value - 0.01, i, f'{value:.3f}', va='center', ha='right', fontsize=9, fontweight='bold')
else:
    ax2.text(0.5, 0.5, 'No Negative Correlations', ha='center', va='center', fontsize=14)
    ax2.set_xticks([])
    ax2.set_yticks([])

plt.tight_layout()
plt.show()

# Statistical summary
print("=" * 60)
print("STATISTICAL SUMMARY OF CORRELATIONS WITH 'HIGH'")
print("=" * 60)
print(f"Mean correlation: {correlation_sorted.mean():.4f}")
print(f"Median correlation: {correlation_sorted.median():.4f}")
print(f"Std deviation: {correlation_sorted.std():.4f}")
print(f"Max correlation: {correlation_sorted.max():.4f}")
print(f"Min correlation: {correlation_sorted.min():.4f}")
print(f"\nFeatures with correlation > 0.9: {(correlation_sorted > 0.9).sum()}")
print(f"Features with correlation > 0.8: {(correlation_sorted > 0.8).sum()}")
print(f"Features with correlation > 0.7: {(correlation_sorted > 0.7).sum()}")
print(f"Features with correlation < 0: {(correlation_sorted < 0).sum()}")
```
